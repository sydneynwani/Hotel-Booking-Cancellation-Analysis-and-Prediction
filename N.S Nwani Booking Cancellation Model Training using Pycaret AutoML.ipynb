```python
from pycaret.classification import * #Importing Relevant Libraries
```

```python
import pandas as pd #Importing relevant library
```

```python
df=pd.read_csv("cleaned_rebalanced_booking_cancellation2b.csv") #Loading the cleaned and preprocessed dataset
```

```python
df.head() #Dataset overview
```

```python
clf_setup=setup(data=df,target="booking status",normalize=True,session_id=42)  #Initializing Pycaret
```

```python
best_model=compare_models(cross_validation=False,sort='f1') #comparing models and selecting the best one
```

```python
print(best_model)
```

```python
tuned_model=tune_model(best_model,optimize="f1",fold=5) #Fine-tuning the best model
```

```python
final_model=finalize_model(tuned_model) #Original model was better than tuned model
```

```python
save_model(final_model,"final_model_booking_cancellation_prediction")  #Saving the final model
```
